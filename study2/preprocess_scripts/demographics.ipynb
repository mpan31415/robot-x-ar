```python
from pandas import read_csv
from numpy import std
import matplotlib.pyplot as plt

NUM_PARTICIPANTS = 24
```

```python
##########################################################################################
def get_raw_data():
    
    my_file_dir = 'c:\\Users\\micha\\OneDrive\\My_GitHub_Repos\\robot-x-ar\\study2\\data\\demographics\\demo_form.csv'
    big_df = read_csv(my_file_dir)
    df = big_df.iloc[2:, 17:]
    
    # print("\n Finished reading raw csv file! \n")
    
    return df
```

```python
demo_df = get_raw_data()

# convert numerical to choice for gender column
demo_df['Q2'] = demo_df['Q2'].apply(lambda x: "male" if int(x)==9 else "female")

# convert pid column to integers
demo_df['Q1'] = demo_df['Q1'].apply(lambda x: int(x))

###### write to new (cleaned) dataframe ######

# write new dataframe to csv file
demo_df.columns = ['pid','gender','age','video_games','mixed_reality','interact_robot','coordination']
demo_df.sort_values(by="pid", inplace=True)
dest_path = 'c:\\Users\\micha\\OneDrive\\My_GitHub_Repos\\robot-x-ar\\study2\\data\\demographics\\demo_cleaned.csv'
demo_df.to_csv(dest_path, index=False)

print(" Successfully written pre-processed data to csv file! \n")
```

```python
def get_descriptive_stats(lst):
    mean = sum(lst)/len(lst)
    stdev = std(lst)
    mini, maxi = min(lst), max(lst)
    print("average = %.3f, stdev = %.3f" % (mean, stdev))
    print("min = %.3f, max = %.3f" % (mini, maxi))
```

```python
gender_list = demo_df['gender'].tolist()
male_count = gender_list.count("male")
female_count = len(gender_list) - male_count
print("# male = %d, # female = %d" % (male_count, female_count))
```

#### 3. Play video games

```python
video_games_list = demo_df['video_games'].apply(lambda x: int(x)).tolist()
get_descriptive_stats(video_games_list)
plt.hist(video_games_list)
plt.title("How often do you play video games?")
plt.show()
```
