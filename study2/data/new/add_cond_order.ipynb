```python
from pandas import read_csv
from os import getcwd
```

```python
##########################################################################################
def get_raw_data():
    my_file_dir = getcwd() + "\\..\\test\\mt_no_outlier_all.csv"
    df = read_csv(my_file_dir)
    return df
```

```python
mt_df = get_raw_data()

mt_df.head()
```

```python
# add column of "cond_order"
# "viz_first" if pid <= 12, "none_first" if pid > 12
mt_df['cond_order'] = mt_df['pid'].apply(lambda x: 'viz_first' if x <= 12 else 'none_first')

# rename "mt" column to "time"
mt_df.rename(columns={'mt': 'time'}, inplace=True)

mt_df
```

```python
# save to new csv file
mt_df.to_csv(getcwd() + "\\mt_data.csv", index=False)
```
