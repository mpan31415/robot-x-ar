```python
import matplotlib.pyplot as plt
from os import getcwd
from pandas import read_csv

%matplotlib inline
```

```python
def get_no_outlier_df():
    my_file_dir = getcwd() + "\\..\\test\\mt_no_outlier_all.csv"
    df = read_csv(my_file_dir)
    return df

def get_no_outlier_by_cond_df():
    my_file_dir = getcwd() + "\\..\\test\\mt_no_outlier_by_cond.csv"
    df = read_csv(my_file_dir)
    return df
```

## 1. Movement Time Altogether

```python
task_df = get_no_outlier_df()
all_mt = task_df['mt'].tolist()

# 2 plots, a boxplot and a violin plot, both showing all data points
plt.figure(figsize=(12, 6))
plt.suptitle('Movement Time (MT) Overall Distribution', fontsize=16)
plt.subplot(1, 2, 1)
plt.boxplot(all_mt, patch_artist=True)
plt.xlabel('MT Values')
plt.ylabel("Movement Time per Trial (s)")
plt.subplot(1, 2, 2)
plt.violinplot(all_mt)
plt.xlabel('MT Values')
plt.ylabel("Movement Time per Trial (s)")
plt.tight_layout()
plt.show()
```

## 2. Movement Time By Condition

```python
task_df = get_no_outlier_df()
yes_arm_mt = task_df[task_df['condition'] == 'yes']['mt'].tolist()
no_arm_mt = task_df[task_df['condition'] == 'no']['mt'].tolist()
assert len(yes_arm_mt) == len(no_arm_mt), "The number of trials in both conditions should be equal."

# for WITH ARM condition
plt.figure(figsize=(12, 6))
plt.suptitle('Movement Time (MT) Distribution (WITH ARM)', fontsize=16)
plt.subplot(1, 2, 1)
plt.boxplot(yes_arm_mt, patch_artist=True)
plt.xlabel('MT Values')
plt.ylabel("Movement Time per Trial (s)")
plt.subplot(1, 2, 2)
plt.violinplot(yes_arm_mt)
plt.xlabel('MT Values')
plt.ylabel("Movement Time per Trial (s)")
plt.tight_layout()
plt.show()

# for WITHOUT ARM condition
plt.figure(figsize=(12, 6))
plt.suptitle('Movement Time (MT) Overall Distribution (NO ARM)', fontsize=16)
plt.subplot(1, 2, 1)
plt.boxplot(no_arm_mt, patch_artist=True)
plt.xlabel('MT Values')
plt.ylabel("Movement Time per Trial (s)")
plt.subplot(1, 2, 2)
plt.violinplot(no_arm_mt)
plt.xlabel('MT Values')
plt.ylabel("Movement Time per Trial (s)")
plt.tight_layout()
plt.show()
```
