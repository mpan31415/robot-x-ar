```python
from pandas import read_csv, DataFrame
from os import getcwd
```

```python
##########################################################################################
def get_task_data():
    my_file_dir = getcwd() + "\\merged_performance_data.csv"
    df = read_csv(my_file_dir)
    return df
```

```python
task_df = get_task_data()
task_df
```

#### Only keep rows with trial # = 2

```python
half_task_df = task_df[task_df['trial']==2]
half_task_df
```

#### save to new csv file

```python
dest_path = getcwd() + "\\half_performance_data.csv"
half_task_df.to_csv(dest_path, index=False)
print(" Successfully written halved task performance data to csv file! \n")
```
