```python
from pandas import read_csv, DataFrame
from os import getcwd
from re import sub
import numpy as np
import matplotlib.pyplot as plt
```

```python
##########################################################################################
def get_cleaned_data():
    
    my_curr_dir = getcwd()
    my_file_dir = my_curr_dir + "\data\questionnaire\main_form_cleaned.csv"
    big_df = read_csv(my_file_dir)
    # num_cols = big_df.shape[1]
    # df = big_df.iloc[:, 17:num_cols]
    # df = df.fillna(1)
    # df.drop(1, axis="index", inplace=True)
    # df.reset_index(drop=True, inplace=True)
    
    print("\n Finished reading raw csv file! \n")
    
    return big_df
```

```python
df = get_cleaned_data()
# grouped_df = df.groupby('condition')
df.groupby(by=['condition']).mean()
```

```python
for i in range(9):
    part_rows = df.iloc[i*4:i*4+4, :].sort_values(by='condition')
    part_tlx_ave_list = part_rows['difficulty'].tolist()
    plt.plot(["A","B","C","D"], part_tlx_ave_list, label='participant '+str(i+1))
# plt.legend()
plt.title("Difficulty")
plt.xlabel('Condition')
plt.ylabel('Difficulty')
plt.show()
```
